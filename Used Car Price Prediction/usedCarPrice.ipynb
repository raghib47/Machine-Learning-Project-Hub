```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
car_dataset = pd.read_csv('train-data.csv')
```

```python
car_dataset
```

```python
car_dataset.info()
```

```python
car_dataset.isnull().sum()
```

```python
car_dataset.drop(["Unnamed: 0","New_Price"],axis=1,inplace=True)
car_dataset
```

```python
car_dataset.isnull().sum()
```

```python
car_dataset = car_dataset.dropna()
car_dataset
```

```python
car_dataset = car_dataset.reset_index()
car_dataset
```

```python
car_dataset.shape
```

```python
for i in range(car_dataset.shape[0]):  
    car_dataset.at[i, 'Company'] = car_dataset['Name'][i].split()[0]
    car_dataset.at[i, 'Mileage(km/kg)'] = car_dataset['Mileage'][i].split()[0]
    car_dataset.at[i, 'Engine(CC)'] = car_dataset['Engine'][i].split()[0]
    car_dataset.at[i, 'Power(bhp)'] = car_dataset['Power'][i].split()[0]
```

```python
car_dataset
```

```python
car_dataset.drop(["index"],axis=1,inplace=True)
car_dataset.drop(["Name"],axis=1,inplace=True)
car_dataset.drop(["Mileage"],axis=1,inplace=True)
car_dataset.drop(["Engine"],axis=1,inplace=True)
car_dataset.drop(["Power"],axis=1,inplace=True)
```

```python
car_dataset
```

```python
car_dataset.info()
```

```python
car_dataset['Mileage(km/kg)'] = car_dataset['Mileage(km/kg)'].astype(float)
car_dataset['Engine(CC)'] = car_dataset['Engine(CC)'].astype(float)
```

```python
car_dataset
```

```python
car_dataset[car_dataset['Power(bhp)']=='null']
```

```python
car_dataset.drop(index = car_dataset[car_dataset['Power(bhp)']=='null'].index, inplace=True)
```

```python
car_dataset
```

```python
car_dataset = car_dataset.reset_index(drop=True)
```

```python
car_dataset
```

```python
car_dataset['Power(bhp)'] = car_dataset['Power(bhp)'].astype(float)
```

```python
car_dataset
```

```python
car_dataset.info()
```

```python
car_dataset['Location'].value_counts()
```

```python
sns.catplot(y='Price',x='Location',data= car_dataset.sort_values('Price',ascending=False),kind="boxen",height=6, aspect=3)
plt.show;
```

### One-hot-encoding here

```python
Location = car_dataset[['Location']]
Location = pd.get_dummies(Location)
Location
```

```python
car_dataset['Fuel_Type'].value_counts()
```

```python
sns.catplot(y='Price',x='Fuel_Type',data= car_dataset.sort_values('Price',ascending=False),kind="boxen",height=6, aspect=3)
plt.show;
```

```python
Fuel_t = car_dataset[['Fuel_Type']]
Fuel_t = pd.get_dummies(Fuel_t)
Fuel_t
```

```python
car_dataset['Transmission'].value_counts()
```

```python
sns.catplot(y='Price',x='Transmission',data= car_dataset.sort_values('Price',ascending=False),kind="boxen",height=6, aspect=3)
plt.show
```

```python
Transmission = car_dataset[['Transmission']]
Transmission = pd.get_dummies(Transmission,drop_first=True)
Transmission
```

```python
car_dataset['Owner_Type'].value_counts()
```

```python
car_dataset['Owner_Type'].replace({"First":4,"Second":3,"Third": 2,"Fourth & Above":1},inplace=True)
car_dataset
```

```python
car_dataset.Company.value_counts()
```

```python
car_dataset.drop(["Company"],axis=1,inplace=True)
```

```python
car_dataset
```

```python
final_train= pd.concat([car_dataset,Location,Fuel_t,Transmission],axis=1)
final_train
```

```python
final_train.drop(["Location","Fuel_Type","Transmission"],axis=1,inplace=True)
final_train
```

```python
price = final_train['Price']
final_train.drop('Price', axis=1,inplace=True)
final_train['Price'] = price
final_train
```

```python
sns.pairplot(car_dataset)
```

# Final Features Selection

```python
X = final_train.drop('Price', axis=1)
```

```python
X
```

```python
y = final_train['Price']
y
```

```python
from sklearn.ensemble import ExtraTreesRegressor
selection= ExtraTreesRegressor()
selection.fit(X,y)
```

```python
plt.figure(figsize = (12,8))
feat_importances = pd.Series(selection.feature_importances_, index=X.columns)
feat_importances.nlargest(20).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
print("Accuracy on Traing set: ",linear_reg.score(X_train,y_train))
print("Accuracy on Testing set: ",linear_reg.score(X_test,y_test))
```

```python
predictions = linear_reg.predict(X_test)
plt.scatter(y_test,predictions);
```

```python
from sklearn import linear_model

lasso_reg = linear_model.Lasso(alpha=40, tol=0.5)
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train, y_train)
```

```python
from sklearn.linear_model import Ridge

ridge_reg= Ridge(alpha=50, tol=0.1)
ridge_reg.fit(X_train, y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor()
dt_reg.fit(X_train, y_train)
```

```python
print("Accuracy on Traing set: ",dt_reg.score(X_train,y_train))
print("Accuracy on Testing set: ",dt_reg.score(X_test,y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor()
rf_reg.fit(X_train, y_train)
```

```python
print("Accuracy on Traing set: ",rf_reg.score(X_train,y_train))
print("Accuracy on Testing set: ",rf_reg.score(X_test,y_test))
```

```python
print("Please enter the following details for the car you want to predict the price for:")


user_year = int(input("Enter the car's manufacturing year (e.g., 2015): "))
user_kilometers_driven = int(input("Enter the Kilometers Driven (e.g., 50000): "))
user_owner_type = int(input("Enter the Owner Type (4=First, 3=Second, 2=Third, 1=Fourth & Above): "))
user_seats = float(input("Enter the number of Seats (e.g., 5.0): "))
user_mileage = float(input("Enter the Mileage in km/kg or kmpl (e.g., 18.5): "))
user_engine = float(input("Enter the Engine capacity in CC (e.g., 1200.0): "))
user_power = float(input("Enter the Power in bhp (e.g., 90.0): "))


user_location = input("Enter the Location (e.g., Mumbai, Chennai, Delhi): ")
user_fuel_type = input("Enter the Fuel Type (e.g., Diesel, Petrol, CNG): ")
user_transmission = input("Enter the Transmission (Manual or Automatic): ")


user_data = {
    'Year': user_year,
    'Kilometers_Driven': user_kilometers_driven,
    'Owner_Type': user_owner_type,
    'Seats': user_seats,
    'Mileage(km/kg)': user_mileage,
    'Engine(CC)': user_engine,
    'Power(bhp)': user_power,
}


user_df = pd.DataFrame(0, index=[0], columns=X_train.columns)


for col, value in user_data.items():
    if col in user_df.columns:
        user_df[col] = value


if f'Location_{user_location}' in user_df.columns:
    user_df[f'Location_{user_location}'] = True
if f'Fuel_Type_{user_fuel_type}' in user_df.columns:
    user_df[f'Fuel_Type_{user_fuel_type}'] = True
if f'Transmission_{user_transmission}' in user_df.columns:
    user_df[f'Transmission_{user_transmission}'] = True


for col in user_df.columns:
    if user_df[col].dtype == 'bool':
        user_df[col] = user_df[col].astype(bool)


predicted_price = rf_reg.predict(user_df)

print(f"\nPredicted Car Price: {predicted_price[0]:.2f} Lakhs")
```
